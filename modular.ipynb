# comparing graphs of individual columns/pointers

```python

for x in range(v1.shape[1]):
    plt.figure(figsize=(7,3.5))
    v1.iloc[:,x].plot(label = "player1"),v2.iloc[:,x].plot(label = "you")
    plt.title(v1.columns[x])
    plt.legend(loc="upper right")
    path = "static/graphs/"+v1.columns[x]+".jpg"
    plt.savefig(path, bbox_inches='tight')
#     plt.show()
```

```python
d = {}
for x in range(v1.shape[1]):
#     print(df.iloc[:,x].corr(js.iloc[:,x]), "of ", df.columns[x])
    d[v1.columns[x]] = v1.iloc[:,x].corr(v2.iloc[:,x])
sum = 0
for x in d.values():
    sum = abs(x)*100 + sum
print("you match player 1 by ",(sum/(v1.shape[1]))+10," %")
```
